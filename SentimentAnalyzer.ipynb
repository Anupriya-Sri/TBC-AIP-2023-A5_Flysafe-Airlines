```python
import os
import numpy as np
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
data = pd.read_excel('Flysafe Airlines.xlsx')
```

```python
data
```

```python
data.describe()
```

```python
review = data['comment']
duplicated_review = data[review.isin(review[review.duplicated()])].sort_values("comment")
duplicated_review.head()
print(review)
```

```python
data.drop_duplicates(inplace = True)
```

```python


# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Sample reviews to analyze
reviews = data['comment']
  
# Analyze each review and print sentiment scores
for review in reviews:
    scores = sia.polarity_scores(review)
   # print(review)
    print(scores)
```
